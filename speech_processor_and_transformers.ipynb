```python

q_a = pipeline('question-answering')

# Summarized context about recursion
context = """
A recursive function is one that calls itself. Without a stopping condition, it leads to an infinite loop and a stack overflow.
Recursion is a different way to create loops but must have a base case to prevent endless calls.

In coding interviews, recursion is often used for problems like computing the Fibonacci sequence.
While it's easy to write recursively, doing so without optimization results in poor performance (O(2^n) time complexity).
This can be improved using memoization.

Recursive functions aren't ideal for everything but are powerful for tree and graph traversal problems.
"""

print("Context about recursion is loaded.")
print("Ask me questions about recursion. Type 'quit' to exit.")

# Interactive question-answering loop
while True:
    question = input("Enter your question: ")
    if question.lower() == 'quit':
        break

    result_txt = q_a({
        'question': question,
        'context': context
    })

    print("Answer:", result_txt['answer'])
    print("-" * 20)

print("Exiting the question-answering session.")
```

```python
from googletrans import Translator
translator = Translator()
english_text = input("Enter text in English: ")
target_lang = input("Enter target language code (e.g., 'fr' for French, 'es' for Spanish): ")

translated = translator.translate(english_text, src='en', dest=target_lang)

print(f"\nTranslated text in {target_lang}: {translated.text}")

```
